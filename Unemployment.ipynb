# Visualization:

Dashboard to analyze the employment rate using bar graph

```python
fg=px.bar(df,x="Region", y='Estimated Employed',color="Region", title ="Employment Rate of India")

fg.show()
```

Dashboard to analyze the unemployment rate  using scatter plot

```python
fg=px.scatter(df,x="Region", y='Estimated Unemployment Rate',color="Region", title ="Unemployment Rate of India")

fg.show()
```

Dashboard to analyze the Estimated Labour Participation Rate using boxplot

```python
fg=px.box(df,x="Region", y='Estimated Labour Participation Rate',color="Region", title ='Estimated Labour Participation Rate')

fg.show()
```

Dashboard to analyze the unemployment rate of each Indian state using sunburst

```python
unemployment=df[["States", "Region","Estimated Unemployment Rate"]]
figure=px.sunburst(unemployment,path=[ "Region","States"],
                   values="Estimated Unemployment Rate",width=700,height=600,color_continuous_scale="RdY1Gn",
                   title=" Unemployment Rate in India")
```

```python
figure.show()
```
