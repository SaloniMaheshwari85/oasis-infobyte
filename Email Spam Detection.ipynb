**Replacing Null Values**

```python
df = df.where((pd.notnull(df)),'')
```

```python
df
```

```python
df = df.rename(columns={'v1' : 'Category','v2' : 'Messages'} )
```

```python
df.head()
```

**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
df['Category'] = label_encoder.fit_transform(df['Category'])
```

**Spliting Data**

```python
x = df['Messages']
y = df['Category']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state= 50, test_size = .2)
```

**Features Extraction**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python

features_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
x_train_features = features_extraction.fit_transform(x_train)
x_test_features = features_extraction.transform(x_test)
```

```python
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

**Model Training**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lo = LogisticRegression()
lo.fit(x_train_features, y_train)
```

**Evaluating**

```python
from sklearn.metrics import accuracy_score
```

```python
prediction = lo.predict(x_train_features)
accuracy = accuracy_score(y_train, prediction)
```

```python
accuracy
```

**Predicting New Value**

```python
mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times."]
features = features_extraction.transform(mail)
```

```python
predictions = lo.predict(features)
print(predictions)

if (predictions[0]==1):
  print('Not a Spam')
else:
  print('Spam')
```
